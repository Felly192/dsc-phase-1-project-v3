```python
# Fill missing categorical data
categorical_fill = ['Airport.Code', 'Airport.Name', 'Injury.Severity', 
                    'Aircraft.damage', 'Registration.Number', 'Amateur.Built', 
                    'Broad.phase.of.flight', 'Report.Status']
for col in categorical_fill:
    df[col].fillna('Unknown', inplace=True)

# Fill numerical data
df['Number.of.Engines'].fillna(df['Number.of.Engines'].median(), inplace=True)

# Drop highly sparse columns if not critical
df.drop(['FAR.Description', 'Schedule'], axis=1, inplace=True)

# Convert dates to datetime format
df['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')
df['Publication.Date'] = pd.to_datetime(df['Publication.Date'], errors='coerce')
```

```python

# Check for unique values and identify inconsistencies
print(df['Country'].unique())
print(df['Location'].unique())

# Standardize and clean 'Country' column
def clean_country(country):
    if pd.isna(country):
        return 'Unknown'
    country = country.strip().title()  # Remove leading/trailing spaces and capitalize
    return country

df['Country'] = df['Country'].apply(clean_country)

# Standardize and clean 'Location' column
def clean_location(location):
    if pd.isna(location):
        return 'Unknown'
    location = location.strip().title()  # Remove leading/trailing spaces and capitalize
    return location

df['Location'] = df['Location'].apply(clean_location)

# Check the results
print(df['Country'].unique())
print(df['Location'].unique())
```

## Visualization
 

 Lets use bar chart for counts for each weather condition

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot
weather_counts = df['Weather.Condition'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=weather_counts.index, y=weather_counts.values, palette='viridis')
plt.title('Weather Condition Distribution')
plt.xlabel('Weather Condition')
plt.ylabel('Count')
plt.show()
```

#### Findings 
The graph depicts the distribution of weather conditions with 4 categories:VMC(Visual Meteorological Conditions),IMC(Instrumental Meteorological Conditions),UNK(Unkown),Unk(Lowercase Unknown).

VMC is the most prevalent weather condition in this dataset i.e it has the highest number of counts while IMC and UNK/Unk make up very small proportions of the total

 Fatal Injuries vs Broad Phase of Flight

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Broad.phase.of.flight', y='Total.Fatal.Injuries', palette='muted',ci =None) # ci=None removes confidence intervals
plt.title('Fatal Injuries by Phase of Flight')
plt.xlabel('Broad Phase of Flight')
plt.ylabel('Total Fatal Injuries')
plt.xticks(rotation=45)
plt.show()
```

#### Findings 
This graphs illustrates fatal injuries across different phases of flight.
1. Unknown and other phases have the highest fatal injuries with their bars  nearly reaching the maximum of of scale 1.0.This suggests that many fatal injuries are either not attributed to specific phases or they occur in 'other'.
2. Approach and maneuvering are phases that exhibit high fatal injuries and they indicate that it might be critical stages of flight that are prone to accidents.
3. cruise and decent have moderate fatal injury counts.
4. Taxi and Landing have the least fatal injuries highlighting their safety compared to other phases


## Accident Rates by Aircraft Type

We can calculate the proportion of accidents for each aircraft type (Make or Model) to understand which types are most commonly involved in accidents.



```python
# Accident rates by Aircraft Type
aircraft_type_counts = df['Make'].value_counts(normalize=True) * 100

# Plot the top 10 most common aircraft types
plt.figure(figsize=(12, 6))
aircraft_type_counts.head(10).plot(kind='bar', color='teal')
plt.title('Accident Rates by Aircraft Type (Top 10)')
plt.xlabel('Aircraft Type')
plt.ylabel('Accident Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

#### Findings

This graph illustrates the accident rates by aircraft type(Top 10),with the accidents rate measured in perecentages.
From our graph we can conclude that:
1. Cessna has the highest accident rate among all aircraft types,
2. Piper has the second highest accident rate while CESSNA and Beech have moderate rates accident ,Bell,Boeing ,Grumman and Mooney have lower rates of accidents with Mooney being the least
3. The dominance of Cessna and Piper in the higher accident rates indicates thats there is need for enhanced safety meausures and pilot training for these aircraft types.
4. Boeing and bell have shown lower accidents rates which may reflect stricter regulations and safety protocals

## Severity of Accidents by Aircraft Type
We will assess how accident severities (fatalities, serious injuries, minor injuries) are distributed across different aircraft types.

```python
# Summarizing severity metrics by aircraft type
severity_by_type = df.groupby('Make')[['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries']].sum()

# Plot the top 5 aircraft types by total fatalities
top_fatalities = severity_by_type.sort_values(by='Total.Fatal.Injuries', ascending=False).head(5)

top_fatalities.plot(kind='bar', figsize=(12, 6), stacked=True, colormap='viridis')
plt.title('Severity of Accidents by Aircraft Type (Top 5)')
plt.xlabel('Aircraft Type')
plt.ylabel('Number of Injuries')
plt.legend(title='Severity')
plt.xticks(rotation=45)
plt.show()
```
